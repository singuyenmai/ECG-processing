```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import shutil

import wfdb
from wfdb import processing

%matplotlib inline
```

## Reading records and annotations

```python
record_id = '16265'
record_path = '../nsrDb/'+record_id
```

```python
# Demo 1 - Read a wfdb record using the 'rdrecord' function into a wfdb.Record object.
# Plot the signals, and show the data.
record = wfdb.rdrecord(record_path) 
annotation = wfdb.rdann(record_path, 'atr')

print(record.__dict__)
```

```python
raw = record.__dict__['p_signal'][:,0] #Take only 1st channel - ECG1
fs = record.fs
print(fs)
print(raw.shape)
print(raw)
```

## Write record into .csv file

```python
time = np.linspace(0, 11730944/fs, 11730944)
print(len(time))
time
```

```python
file = pd.DataFrame()
file['timestamp'] = time
file['ecg_measurement'] = raw
file
```

```python
file_name = '../tmp/nsrDb_' + record_id + '.csv'
file.to_csv(file_name)
```
