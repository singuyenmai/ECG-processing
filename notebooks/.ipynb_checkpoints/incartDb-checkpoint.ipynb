# Run all samples - findpeaks_limit = 0.1 & findpeaks_spacing_factor=0.2

```python
summary = pd.DataFrame()

for record_id in record_ids:
    record_path = database_dir + record_id
    
    print("Analyzing record {:s} ...".format(record_id))
    
    # Run QRSeeker
    qrs_seeker = QRSeeker(record_path, channels=[0,9],
                          findpeaks_limit=0.1, findpeaks_spacing_factor=0.2, 
                          verbose=False) 
    
    # Store the results
    results = {
        'record_id': record_id,
        
        'ref_annotation': qrs_seeker.ref_annotation,
        'test_annotation': np.array(qrs_seeker.detected_inds),
        
        'matched_ref_inds': qrs_seeker.matched_ref_inds,
        'matched_test_inds': qrs_seeker.matched_test_inds,
        'unmatched_ref_inds': qrs_seeker.unmatched_ref_inds,
        'unmatched_test_inds': qrs_seeker.unmatched_test_inds,

        'true_positive': qrs_seeker.true_positive,
        'false_positive': qrs_seeker.false_positive,
        'false_negative': qrs_seeker.false_negative,

        'specificity': qrs_seeker.specificity,
        'positive_predictivity': qrs_seeker.positive_predictivity,
        'FPR': qrs_seeker.fpr
    }
    
    summary = pd.concat([summary, pd.DataFrame([results], columns = results.keys())])
    print("\n")
```

```python
summary.head()
```

```python
summary[['specificity', 'positive_predictivity', 'FPR']].describe()
```

**There were only 20 samples with well performed QRS detection** <br>
**Setting threshold for number of peaks detected to 85% limited the specificity to be minimum of 85%**

```python
summary.to_csv("incartDb_summary.limit0.1_spacing0.2.csv", sep="\t", index=False)
```

# Check bad runs and re-analyse samples with different parameters (optimizing)

### Findpeaks_limit = 0.01

```python
todo_record_ids = summary.loc[summary['specificity'].isna() == True, 'record_id']

len(todo_record_ids)
```

```python
summary_newrun = pd.DataFrame()

for record_id in todo_record_ids:
    record_path = database_dir + record_id
    
    print("Analyzing record {:s} ...".format(record_id))
    
    # Run QRSeeker
    qrs_seeker = QRSeeker(record_path, channels=[0,9],
                          findpeaks_limit=0.01, findpeaks_spacing_factor=0.2, 
                          verbose=False) 
    
    # Store the results
    results = {
        'record_id': record_id,
        
        'ref_annotation': qrs_seeker.ref_annotation,
        'test_annotation': np.array(qrs_seeker.detected_inds),
        
        'matched_ref_inds': qrs_seeker.matched_ref_inds,
        'matched_test_inds': qrs_seeker.matched_test_inds,
        'unmatched_ref_inds': qrs_seeker.unmatched_ref_inds,
        'unmatched_test_inds': qrs_seeker.unmatched_test_inds,

        'true_positive': qrs_seeker.true_positive,
        'false_positive': qrs_seeker.false_positive,
        'false_negative': qrs_seeker.false_negative,

        'specificity': qrs_seeker.specificity,
        'positive_predictivity': qrs_seeker.positive_predictivity,
        'FPR': qrs_seeker.fpr
    }
    
    summary_newrun = pd.concat([summary_newrun, pd.DataFrame([results], columns = results.keys())])
    print("\n")
```

```python
summary_newrun[['specificity', 'positive_predictivity', 'FPR']].describe()
```

```python
pd.concat([summary_newrun[['specificity', 'positive_predictivity', 'FPR']], 
           summary[['specificity', 'positive_predictivity', 'FPR']]]).describe()
```

```python
sobad_record_ids = summary_newrun.loc[summary_newrun['specificity'].isna() == True, 'record_id']

detected_lengths = [len(x) for x in list(summary_newrun.loc[summary_newrun['specificity'].isna() == True, 'test_annotation'])]
ref_lengths = [len(x) for x in list(summary_newrun.loc[summary_newrun['specificity'].isna() == True, 'ref_annotation'])]

print(list(sobad_record_ids))
print(np.array(detected_lengths)*100 / np.array(ref_lengths))
```

### Findpeaks_limit = None

```python
summary_supernewrun = pd.DataFrame()

for record_id in sobad_record_ids:
    record_path = database_dir + record_id
    
    print("Analyzing record {:s} ...".format(record_id))
    
    # Run QRSeeker
    qrs_seeker = QRSeeker(record_path, channels=[0,9],
                          findpeaks_limit=None, findpeaks_spacing_factor=0.2, 
                          verbose=False) 
    
    # Store the results
    results = {
        'record_id': record_id,
        
        'ref_annotation': qrs_seeker.ref_annotation,
        'test_annotation': np.array(qrs_seeker.detected_inds),
        
        'matched_ref_inds': qrs_seeker.matched_ref_inds,
        'matched_test_inds': qrs_seeker.matched_test_inds,
        'unmatched_ref_inds': qrs_seeker.unmatched_ref_inds,
        'unmatched_test_inds': qrs_seeker.unmatched_test_inds,

        'true_positive': qrs_seeker.true_positive,
        'false_positive': qrs_seeker.false_positive,
        'false_negative': qrs_seeker.false_negative,

        'specificity': qrs_seeker.specificity,
        'positive_predictivity': qrs_seeker.positive_predictivity,
        'FPR': qrs_seeker.fpr
    }
    
    summary_supernewrun = pd.concat([summary_supernewrun, pd.DataFrame([results], columns = results.keys())])
    print("\n")
```

```python
detected_lengths = [len(x) for x in list(summary_supernewrun.loc[summary_supernewrun['specificity'].isna() == True, 'test_annotation'])]
ref_lengths = [len(x) for x in list(summary_supernewrun.loc[summary_supernewrun['specificity'].isna() == True, 'ref_annotation'])]

print(np.array(detected_lengths)*100 / np.array(ref_lengths))
```

```python
summary_supernewrun[['specificity', 'positive_predictivity', 'FPR']].describe()
```

# Create final summary table - 55 good samples

```python
df1 = summary[summary['specificity'].isna() == False].copy()
df2 = summary_newrun[summary_newrun['specificity'].isna() == False].copy()
df3 = summary_supernewrun[summary_supernewrun['specificity'].isna() == False].copy()

summary_final = pd.concat([df1, df2, df3])
summary_final.reset_index(drop=True, inplace=True)
```
