````python
import pickle

import wfdb
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import dabest

%matplotlib inline
````

All records of IncartDb have sampling frequency of 257 Hz (as stated on their website) <br>
All records of NsrDb have sampling frequency of 128 Hz (checked by the below code)

```python
import wfdb

database="nsrdb"
database_dir = '../nsrDb/'
record_ids = wfdb.get_record_list(database)

for id in record_ids:                                                       
    record_path = database_dir + id                                         
    fs = wfdb.rdrecord(record_path).__dict__['fs']                          
    print(fs) 
```

````python
def HR_estimating(qrs_inds, fs):
    d = np.ediff1d(qrs_inds) # distances in # of samples between consecutive detected peaks
    D = d/fs # distances in second between consecutive detected peaks
    HR = np.mean(60/D) # heart rate in beat per minute
    return [D, HR]    
````

# IncartDb

````python
incart = pickle.load(open("incartDb_summary.pickle", "rb"))

print(incart.shape)
incart.head()
````

````python
heart_rate1 = incart[['record_id']].copy()
heart_rate1['class'] = 'Diseased'
````

````python
patients = set()
for i in heart_rate1['record_id']:
    meta = wfdb.rdrecord('../incartDb/'+i).__dict__['comments']
    patients.add(meta[1])
    
print(patients)
len(patients)
````

````python
heart_rate1['Sex'] = heart_rate1['record_id'].apply(
                                                lambda x: 'Female' if 
                                                wfdb.rdrecord('../incartDb/'+x).__dict__['comments'][0].split(" ")[3]=='F'
                                                else 'Male')
````

````python
heart_rate1['D'] = incart['test_annotation'].apply(lambda x: HR_estimating(x, 257)[0])
heart_rate1['HR'] = incart['test_annotation'].apply(lambda x: HR_estimating(x, 257)[1])

heart_rate1
````

````python
heart_rate1['HR'].describe()
````

# NsrDb

````python
nsr = pickle.load(open("nsrDb_summary.pickle", "rb"))

print(nsr.shape)
nsr.head()
````

````python
N = 128*30*60 # Number of samples in 30 min

heart_rate2 = nsr[['record_id']].copy().astype(str)
heart_rate2['class'] = 'Healthy'

heart_rate2['Sex'] = heart_rate2['record_id'].apply(
                                                lambda x: 'Female' if 
                                                wfdb.rdrecord('../nsrDb/'+x).__dict__['comments'][0].split(" ")[1]=='F'
                                                else 'Male')

heart_rate2['D'] = nsr['test_annotation'].apply(lambda x: HR_estimating(x[x <= N], 128)[0])
heart_rate2['HR'] = nsr['test_annotation'].apply(lambda x: HR_estimating(x[x <= N], 128)[1])

heart_rate2
````

````python
heart_rate2['HR'].describe()
````

# Big table and Statistical Analysis

````python
heart_rate = pd.concat([heart_rate1, heart_rate2])

heart_rate
````

````python
fig1, stat1 = dabest.plot(heart_rate, x='class', y='HR', idx=('Healthy','Diseased'), color_col='Sex')

print(stat1)
````
